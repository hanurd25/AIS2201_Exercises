```python
import numpy as np
import matplotlib.pyplot as plt
M = 37
n = np.arange(M)
w_c = 0.3*np.pi

hn = w_c*np.sinc((n-(M-1)/2)*w_c/np.pi)/np.pi
plt.close(2), plt.figure(2)
plt.stem(hn)


#displayFreqencyResponse(hn, num= 2)
#For displaying frequency response
```

```python
#kodeeksempel 2
import scipy.signal as sig
M = 37
n = np.arange(M)

w_c = 0.4*np.pi

hn_rect = w_c/np.pi*np.sinc(w_c/np.pi*(n-(M-1)/2))
hn_hamming = hn_rect *sig.windows.hamming(M)

hn_lp = w_c/np.pi*np.sinc(0.3*((n-M)/2))
hn_lp = sig.windows.hamming(M)

hn_hp = hn_lp*2*np.cos(0.4*np.pi*(n-(M-1)/2))

plt.close(2), plt.figure(2)
plt.stem(hn_hamming)

plt.close(2), plt.figure(2)
plt.stem(hn_hp)
```
