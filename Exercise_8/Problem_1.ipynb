### First order difference filter

$$y[n] = x[n] -x[n-1] \tag{1}$$

### Signal smoothing and Moving Average Filters

While difference filtering intensity profiles may highlight edges of objects, it may also highlight small localized image details or simple noise. To avoid this, it may be advantageous to combine the difference filter with a smoothing filter (a.k.a. lowpass filter). One of the simplest lowpass FIR filters is the Moving Average filter which calculates the average value of the $M$ most recent signal samples:

$$y[n] = \frac{1}{M} \sum_{k=0}^{M-1} x[n-k] \tag{2}$$

## d)

Use a 4-point moving average filter to conduct smoothing of the intensity profile $x[n]$, and plot the smoothed profile (let's call it $y_s[n]$) alongside the original.

```python
# WRITE YOUR CODE IN THIS CELL:
# inspiration: https://www.geeksforgeeks.org/how-to-calculate-moving-averages-in-python/

#Window size will be 4 for 4 point moving average

x = intensity_profile

def movingAvg(signalArray, windowSize):
    i = 0
    movingAverages = []
    while i < len(signalArray) - windowSize + 1:
        window = signalArray[i : i + windowSize]
        windowAverage = round(sum(window) / windowSize, 2)
        movingAverages.append(windowAverage)
        i += 1
    return movingAverages

plt.figure(figsize=(10, 6))
plt.plot(movingAvg(x, 4), color="green", label="4-Point Moving Average of x")
plt.plot(x, color="red", label="signal x")

plt.xlabel("x-values")
plt.ylabel("Filteret vs. un-filtered itensity")

plt.legend(loc='best') #can't see labels without this
plt.show()
```

### Cascade FIR filters

The process of smoothing the intensity profile $x[n]$ using a moving average filter $h_{s}[n]$ before highlighting edges with a difference filter $h_d[n]$ can be described mathematically as follows:
$$y[n] = (x[n]*h_s[n])*h_d[n]$$

Since convolution is associative this can be simplified into one filtering operation $y[n] = x[n]*h[n]$ where $h[n] = h_s[n]*h_d[n]$.

```python
#https://medium.com/@rohit-krishna/coding-canny-edge-detection-algorithm-from-scratch-in-python-232e1fdceac7
#https://towardsdatascience.com/canny-edge-detection-step-by-step-in-python-computer-vision-b49c3a2d8123
#https://scikit-image.org/skimage-tutorials/lectures/1_image_filters.html
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

Im = plt.imread("data/coins.png")

def defferenceFilterOutput(signalX):
    yVals = []
    for n in range(len(signalX)):
        yVals.append(signalX[n] -signalX[n-1])
    return yVals

#Replaces every color with itensity under a certain treshold with white
def tresholdFilter(signalX, threshold, colorWhite):
    signalReplic = signalX
    rows, cols = signalReplic.shape
    for n in range(rows):
        for i in range(cols):
            if signalReplic[n,i] < threshold:
                signalReplic[n,i] = colorWhite
    return signalReplic

horizontalPic = np.array([defferenceFilterOutput(row) for row in Im])
verticalPic = np.array([defferenceFilterOutput(col) for col in Im.T]).T
YCombined = np.sqrt(horizontalPic**2 + verticalPic**2)

itensityProfile = 140


print(YCombined[itensityProfile])

plt.figure(figsize=(10, 6))
plt.imshow(YCombined, cmap="gray")
rows, cols = Im.shape
plt.plot([0, cols-1], [140, 140], 'r')
plt.title("Edge Detection")
plt.colorbar()
plt.show()


plt.figure(figsize=(10, 6))
plt.imshow(tresholdFilter(YCombined, 0.25, 0.00392157), cmap="gray")
rows, cols = Im.shape
plt.plot([0, cols-1], [140, 140], 'r')
plt.title("Edge Detection")
plt.colorbar()
plt.show()
```
