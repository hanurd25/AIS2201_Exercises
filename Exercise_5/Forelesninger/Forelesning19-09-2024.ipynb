```python
#Regneeksempel 1
import numpy as np 
from math import pi
import matplotlib.pyplot as plt

N = 256 # antall sampler

fs = 4000 #hz
t = np.arange(N)/fs
#dft med n 256 #sampl#er



xt = np.cos(2*pi*570*t)
frekvensopplosning = 4000/256 #FS/n

m = 570/15.625

fm = m * frekvensopplosning

m = 570/ frekvensopplosning #faktisk frekvens

#VindusFrekvens
f = 36*15,525 # = 562.5
plt.figure()
plt.plot(np.abs(np.fft.fft(xt)))
```

```python
#kodeeksempel 2


N = 32
n = np.arange(N)
xn = np.cos(pi/10*n)

wn = 0.5 - 0.5*np.cos(2*pi*n/N)

xn_win= xn*wn

Xn = np.fft.fft(xn)
xm_win  = np.fft.fft(xn_win)

plt.figure()
plt.stem(np.abs(Xn),markerfmt=".", label="uten vindu")
plt.stem(np.abs(xm_win), markerfmt="xC1", linefmt="C1", label="mad vindu")
plt.legend()
```

```python
#kodeeksempel 3


N = 32
fs = 32
padding_factor = 2

f_padded = np.arange(padding_factor*N)/fs/padding_factor


f_n = np.arange(N)/fs
xn = np.cos(pi/10*n)

wn = 0.5 - 0.5*np.cos(2*pi*n/N)



xn_win= xn*wn

xn_padded = np.zeros(N*padding_factor)
xn_padded[:N] = xn_win

Xm = np.fft.fft(xn_win)
Xm_padded = np.fft.fft(xn_padded)

plt.figure(figsize=(12,8))
plt.stem(f_n,np.abs(Xm))
#plt.stem(np.arange(2*N)/fs/2,np.abs(Xm_padded), linefmt="C1", label="")
plt.plot(f_padded,np.abs(Xm_padded),"C1", color="blue", label="")
```
